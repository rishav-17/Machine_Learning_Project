## Importing Libraries

```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
```

## Reading the CSV Document

```python
df= pd.read_csv("C:\\Users\\goris\\OneDrive\\Documents\\Whole training of ml project\\general_diseases\\Training.csv")
```

```python
df.head()
```

```python
df.info()
```

## Encodingand labeling the data


```python
encoder= LabelEncoder()
```

```python
prog=df.prognosis.value_counts()
prog
```

## Creating an empty Label file

```python
df["label"]=prog
df.head()
```

```python
df.label= encoder.fit_transform(df.prognosis)
```

```python
countt=df.label.value_counts()
countt
```

```python
df.head()
```

## Selecting the independent variables

```python
x= df.iloc[:,1:33]
x.head()
```

## Selecting the dependent variable

```python
y= df.iloc[:,-1]
y
```

## Cetagorizing the data into train and test

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.20, shuffle=True)
```

## Defining the Algorithm

```python
model=RandomForestClassifier(n_estimators=50, criterion='log_loss')
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

## Training the model

```python
model.fit(x_train,y_train)
```

## Getting the predictions

```python
y_pred=model.predict(x_test)
```

## Displaying the Accuracy

```python
accu=accuracy_score(y_test,y_pred)
score=accu*100
formatted="{:.3f}".format(score)
print(f"The model is {formatted}% accurate")
```

## Classification report

```python
print(classification_report(y_test,y_pred))
```

## Testing the model on provided Data
